# Exercise 05 : Boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Analizing variancies using boxplot.

```python
df = pd.read_csv('../data/ab-test.csv')

fig, axs = plt.subplots(1, 2, figsize=(15, 8))

df_after = df[df['time'] == 'after']
df_before = df[df['time'] == 'before']

df_after.boxplot(column='diff', by='group', ax=axs[0], widths=0.7, boxprops=dict(linewidth=3, color='green'), 
                 medianprops=dict(linewidth=2, color='red'), whiskerprops=dict(linewidth=1, color='red'), 
                 capprops=dict(linewidth=1, color='red'))
axs[0].set_title('after', fontsize=15)
axs[0].set_xlabel('Group')
axs[0].set_ylabel('Difference (hours)')

df_before.boxplot(column='diff', by='group', ax=axs[1], widths=0.7, boxprops=dict(linewidth=3, color='green'), 
                  medianprops=dict(linewidth=2, color='red'), whiskerprops=dict(linewidth=1, color='red'), 
                  capprops=dict(linewidth=1, color='red'))
axs[1].set_title('before', fontsize=15)
axs[1].set_xlabel('Group')

plt.suptitle('Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
for ax in axs:
    ax.grid(True)

plt.tight_layout()
plt.show()
```

## What was the IQR of the control group before the newsfeed?

```python
control_before = df[(df['group'] == 'control') & (df['time'] == 'before')]

Q1 = control_before['diff'].quantile(0.25)
Q3 = control_before['diff'].quantile(0.75)

IQR = Q3 - Q1
print(f"IQR for control group: {IQR}")
```
